```python
import pandas as pd
pd.set_option('display.max_rows', None)
# set the max columns to none
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_parquet('/tmp/example_2.parquet')
df = df[['CODE','SHARE NAME', 	'INDUSTRY', 	'SHARES *', 	'WEIGHT']]

# Change type of WEIGHT to FLOAT and strip %
df['WEIGHT'] = df['WEIGHT'].str.rstrip('%').astype(float)
result_df = df.groupby('fund', group_keys=False).apply(lambda x: x.sort_values('WEIGHT', ascending=False))
```

```python

# Get the first 10 rows for each 'fund'
top_10_per_fund = result_df.groupby('fund').head(5)
#top_10_per_fund

# Find SHARE NAMEs that appear in multiple funds
#duplicate_share_names = top_10_per_fund[top_10_per_fund.duplicated(subset='SHARE NAME', keep=False)]
duplicate_share_names = df[df.duplicated(subset='SHARE NAME', keep=False)]

# Get the count of occurrences for each duplicated SHARE NAME
share_name_counts = duplicate_share_names['SHARE NAME'].value_counts()
share_name_counts

#mtn_occurrences = df[df['SHARE NAME'] == 'MTN GROUP']
#mtn_occurrences
```
